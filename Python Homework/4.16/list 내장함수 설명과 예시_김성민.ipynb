**list 내장함수 설명과 예시**

---

append(x)	리스트 맨 끝에 원소 x 추가.


---


extend(iterable)	다른 iterable의 모든 요소를 리스트에 추가.


---


insert(i, x)	i 위치에 원소 x 삽입.


---


remove(x)	리스트에서 첫 번째로 나오는 x 삭제. 없으면 오류.


---


pop(i)	인덱스 i의 원소 제거하고 반환. 인덱스 없으면 마지막 원소 제거.


---


clear()	리스트 모든 요소 제거.


---


index(x)	x의 첫 번째 위치(인덱스) 반환. 없으면 오류.


---


count(x)	리스트에서 x의 개수 반환.


---


sort()	리스트 원소 오름차순 정렬.


---


reverse()	리스트 원소 순서를 반대로 뒤집음.


---


copy()	리스트 얕은 복사본 반환.


```python
# append
lst = [1, 2]
lst.append(3)
print(lst)  # [1, 2, 3]
```

```python
# extend
lst.extend([4, 5])
print(lst)  # [1, 2, 3, 4, 5]
```

```python
# insert
lst.insert(2, 99)
print(lst)  # [1, 2, 99, 3, 4, 5]
```

```python
# remove
lst.remove(99)
print(lst)  # [1, 2, 3, 4, 5]
```

```python
# pop
x = lst.pop(1)
print(x, lst)  # 2 [1, 3, 4, 5]
```

```python
# clear
lst.clear()
print(lst)  # []
```

```python
# index
lst = [1, 2, 3, 2]
print(lst.index(2))  # 1
```

```python
# count
print(lst.count(2))  # 2
```

```python
# sort
lst = [5, 1, 3]
lst.sort()
print(lst)  # [1, 3, 5]
```

```python
# reverse
lst.reverse()
print(lst)  # [5, 3, 1]
```

```python
# copy
lst2 = lst.copy()
print(lst2)  # [5, 3, 1]
```
