**random 모듈 설명과 예시**


---

random()	0.0 이상 1.0 미만의 난수(float) 반환.


---


uniform(a, b)	a 이상 b 이하의 float 난수 반환.


---


randint(a, b)	a 이상 b 이하의 정수 난수 반환.


---


randrange(start, stop)	start 이상 stop 미만의 정수 반환.


---


choice(seq)	시퀀스에서 임의의 원소 하나 반환.


---


choices(seq, k=n)	시퀀스에서 중복 허용하며 n개 선택.


---


sample(seq, k=n)	시퀀스에서 중복 없이 n개 선택.


---


shuffle(seq)	시퀀스의 요소 순서를 무작위로 섞음 (리스트만 가능).


---


seed(x)	난수 생성 초기값 설정 (재현 가능하도록).

```python
import random

# random()
print(random.random())  # 예: 0.6745
```

```python
# uniform
print(random.uniform(1.5, 5.5))  # 예: 3.82
```

```python
# randint
print(random.randint(1, 10))  # 1~10 중 랜덤 정수
```

```python
# randrange
print(random.randrange(1, 10))  # 1~9 중 랜덤 정수
```

```python
# choice
print(random.choice(['a', 'b', 'c']))  # 예: 'b'
```

```python
# choices
print(random.choices(['a', 'b', 'c'], k=2))  # 예: ['a', 'c']
```

```python
# sample
print(random.sample([1, 2, 3, 4], 2))  # 예: [3, 1] (중복 없음)
```

```python
# shuffle
lst = [1, 2, 3, 4]
random.shuffle(lst)
print(lst)  # 예: [3, 1, 4, 2]
```

```python
# seed
random.seed(42)
print(random.random())  # 항상 같은 값
```
