**tuple 내장함수 설명과 예시**


---

count(x)	원소 x가 튜플에 몇 번 등장하는지 반환.

---


index(x[, start[, end]])	x가 처음 나타나는 인덱스를 반환. 범위 지정 가능.

---


len(t)	튜플의 전체 원소 개수 반환.

---


in, not in	특정 원소가 튜플에 있는지 확인. (True/False) 반환.

---


"+"	두 튜플을 이어붙여 새 튜플 반환.

---


"*"	튜플을 반복하여 새 튜플 생성.

---


sorted(t)	튜플을 정렬한 리스트 반환 (튜플은 정렬되지 않지만 리스트로 변환됨).

---


max(t) / min(t)	튜플에서 가장 큰 값 / 가장 작은 값 반환.

---


sum(t)	튜플의 모든 숫자 원소의 합 반환.

---


any(t) / all(t)	any: 하나라도 참이면 True, all: 모두 참이어야 True.


---


tuple(iterable)	반복 가능한 객체를 튜플로 변환.

```python
t = (1, 2, 2, 3, 4)

# count
print(t.count(2))  # 2
```

```python
# index
print(t.index(3))  # 3
print(t.index(2, 2))  # 2 (2번째 인덱스부터 검색)
```

```python
# len
print(len(t))  # 5
```

```python
# in, not in
print(2 in t)      # True
print(5 not in t)  # True
```

```python
# +
t2 = (5, 6)
print(t + t2)  # (1, 2, 2, 3, 4, 5, 6)
```

```python
# *
print(t * 2)  # (1, 2, 2, 3, 4, 1, 2, 2, 3, 4)
```

```python
# sorted
print(sorted(t))  # [1, 2, 2, 3, 4] (리스트로 반환됨)
```

```python
# max, min
print(max(t))  # 4
print(min(t))  # 1
```

```python
# sum
print(sum(t))  # 12
```

```python
# any, all
bool_tup = (0, False, 3)
print(any(bool_tup))  # True (3이 참이라서)
print(all(bool_tup))  # False (0과 False가 있어서)
```

```python
# tuple()
lst = [1, 2, 3]
tup = tuple(lst)
print(tup)  # (1, 2, 3)
```
