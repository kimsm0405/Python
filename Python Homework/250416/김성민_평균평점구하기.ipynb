```python
import random

# 학점별 점수 변환표
grade_point = {
    "A+": 4.5, "A": 4.0, "B+": 3.5, "B": 3.0,
    "C+": 2.5, "C": 2.0, "D+": 1.5, "D": 1.0,
    "F": 0.0
}
```

```python
# 1. 성적 데이터 생성
def generate_grades():
    subjects = [chr(97 + i) for i in range(20)]  # a ~ t (20개 과목)
    data = []

    for subject in subjects:
        credit = random.randint(1, 3)  # 학점 1~3
        score = random.randint(50, 100)  # 점수 50~100
        if score >= 95:
            grade = "A+"
        elif score >= 90:
            grade = "A"
        elif score >= 85:
            grade = "B+"
        elif score >= 80:
            grade = "B"
        elif score >= 75:
            grade = "C+"
        elif score >= 70:
            grade = "C"
        elif score >= 65:
            grade = "D+"
        elif score >= 60:
            grade = "D"
        else:
            grade = "F"
        data.append((subject, credit, score, grade))

    return data
```

```python
# 2. txt 파일로 저장
def save_to_file(data, filename="grades.txt"):
    with open(filename, "w", encoding="utf-8") as f:
        for item in data:
            f.write(f"{item[0]} {item[1]} {item[2]} {item[3]}\n")
```

```python
# 3. txt 파일 불러오기
def load_from_file(filename="grades.txt"):
    data = []
    with open(filename, "r", encoding="utf-8") as f:
        for line in f:
            subject, credit, score, grade = line.strip().split()
            data.append((subject, int(credit), int(score), grade))
    return data
```

```python
# 4. 평균평점 계산
def calculate_gpa(data):
    total_points = 0
    total_credits = 0
    for subject, credit, score, grade in data:
        total_credits += credit
        total_points += grade_point[grade] * credit
    gpa = total_points / total_credits
    return round(gpa, 2)
```

```python
# 5. 결과 파일에 GPA 추가 저장
def add_gpa_to_file(gpa, filename="grades.txt"):
    with open(filename, "a", encoding="utf-8") as f:
        f.write(f"\n평균 평점(GPA): {gpa}")
```

```python
# 실행 흐름
def run_program():
    data = generate_grades()
    save_to_file(data)
    loaded_data = load_from_file()
    gpa = calculate_gpa(loaded_data)
    add_gpa_to_file(gpa)
    print(f"평균 평점은 {gpa}입니다. (grades.txt 파일에 저장됨)")
```

```python
# 프로그램 실행
run_program()
```
