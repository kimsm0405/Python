```python
def get_scores(player_name):
    print(f"\n{player_name}님의 점수를 입력하세요.")
    scores = []

    for frame in range(1, 11):
        print(f"\n{frame}프레임:")
        while True:
            try:
                first = int(input(" - 첫 번째 투구 (0~10): "))
                if 0 <= first <= 10:
                    break
                else:
                    print("0~10 사이 숫자만 입력하세요.")
            except ValueError:
                print("숫자를 입력해주세요.")

        while True:
            try:
                max_second = 10 - first
                second = int(input(f" - 두 번째 투구 (0~{max_second}): "))
                if 0 <= second <= max_second:
                    break
                else:
                    print(f"0~{max_second} 사이 숫자만 입력하세요.")
            except ValueError:
                print("숫자를 입력해주세요.")

        scores.append([first, second])  # 두 점수를 리스트로 저장

    return scores
```

```python
# 플레이어 정보
players = []
scores_dict = {}
total_dict = {}
average_dict = {}

print(" 볼링 게임 시작! (5명 참가)")
```

```python
# 이름 입력 및 점수 저장
for i in range(5):
    name = input(f"\n플레이어 {i+1}의 이름을 입력하세요: ")
    players.append(name)
    scores = get_scores(name)
    scores_dict[name] = scores
    total = sum([sum(frame) for frame in scores])  # 각 프레임 점수 합산
    total_dict[name] = total
    average_dict[name] = round(total / 10, 2)  # 평균 계산 (소수 둘째 자리)

# 승자 계산
winner = max(total_dict, key=total_dict.get)
```

```python
# 결과 출력
print("\n 게임 결과:")
for name in players:
    print(f"\n{name} 점수:")
    for i, frame in enumerate(scores_dict[name]):
        print(f" {i+1}프레임: {frame[0]} / {frame[1]}")
    print(f"  총점: {total_dict[name]}점")
    print(f"  평균: {average_dict[name]}점")

print(f"\n 승자: {winner}님!")
```

```python
# 결과 파일 저장
filename = "볼링_게임_결과.txt"
f = open(filename, "w", encoding="utf-8")

f.write(" 볼링 게임 결과 (5인)\n\n")
for name in players:
    f.write(f"{name} 점수:\n")
    for i, frame in enumerate(scores_dict[name]):
        f.write(f" {i+1}프레임: {frame[0]} / {frame[1]}\n")
    f.write(f" 총점: {total_dict[name]}점\n")
    f.write(f" 평균: {average_dict[name]}점\n\n")

f.write(f" 승자: {winner}님!\n")
f.close()

print(f"\n 결과가 '{filename}' 파일에 저장되었습니다!")
```
