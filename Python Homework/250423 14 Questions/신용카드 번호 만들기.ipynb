```python
def is_valid_card(card):
    digits = card.replace("-", "")

    if digits[0] not in "456":
        return False

    if len(digits) != 16:
        return False

    if not digits.isdigit():
        return False

    for ch in card:
        if not (ch.isdigit() or ch == "-"):
            return False

    if "-" in card:
        parts = card.split("-")
        if len(parts) != 4 or any(len(p) != 4 for p in parts):
            return False

    count = 1
    for i in range(1, len(digits)):
        if digits[i] == digits[i - 1]:
            count += 1
            if count >= 4:
                return False
        else:
            count = 1

    return True

def validate_cards():
    n = int(input("입력할 카드 번호 개수를 입력하세요: "))
    for _ in range(n):
        card = input("카드 번호를 입력하세요: ")
        if is_valid_card(card):
            print("Valid")
        else:
            print("Invalid")

validate_cards()
```
