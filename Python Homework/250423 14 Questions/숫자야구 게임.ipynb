```python
import random

def generate_number():
    return [random.randint(0, 9) for _ in range(3)]

def get_feedback(answer, guess):
    strike = 0
    ball = 0
    for i in range(3):
        if guess[i] == answer[i]:
            strike += 1
        elif guess[i] in answer:
            ball += 1
    if strike == 0 and ball == 0:
        return "Out"
    return f"{ball}Ball {strike}Strike"

def play_number_baseball():
    answer = generate_number()
    tries = 0
    print("숫자야구 게임을 시작합니다! (세 자리 숫자를 입력하세요. 예: 123)")

    while True:
        user_input = input("숫자를 입력하세요: ")
        if len(user_input) != 3 or not user_input.isdigit():
            print("세 자리 숫자를 정확히 입력해주세요.")
            continue

        guess = [int(d) for d in user_input]
        tries += 1
        result = get_feedback(answer, guess)
        print("결과:", result)

        if result.endswith("3Strike"):
            print(f"정답입니다! {tries}번 만에 맞췄어요!")
            break

play_number_baseball()
```
