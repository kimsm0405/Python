**dict 내장함수 설명과 예시**


---


get(key, default)	키에 해당하는 값을 반환. 키가 없으면 default 반환.


---


keys()	딕셔너리의 모든 키를 반환.


---


values()	딕셔너리의 모든 값을 반환.


---


items()	(키, 값) 쌍의 튜플들을 반환.


---


pop(key, default)	키의 값을 반환하고 항목 삭제. 키 없으면 default 반환.


---


popitem()	마지막 (키, 값) 쌍을 반환하고 제거.


---


update(other_dict)	다른 딕셔너리의 내용을 추가하거나 기존 키 업데이트.


---


clear()	모든 항목 삭제.


---


copy()	얕은 복사본 반환.


---


fromkeys(seq, value)	주어진 키 목록으로 새 딕셔너리 생성. 값은 기본값으로 설정됨.


---


setdefault(key, default)	키가 없으면 추가하고, 값을 반환.

```python
# get
d = {'a': 1, 'b': 2}
print(d.get('a'))         # 1
print(d.get('c', 0))      # 0
```

```python
# keys, values, items
print(d.keys())           # dict_keys(['a', 'b'])
print(d.values())         # dict_values([1, 2])
print(d.items())          # dict_items([('a', 1), ('b', 2)])
```

```python
# pop
print(d.pop('b'))         # 2
print(d)                  # {'a': 1}
```

```python
# popitem
d = {'a': 1, 'b': 2}
print(d.popitem())        # ('b', 2)
print(d)                  # {'a': 1}
```

```python
# update
d.update({'c': 3})
print(d)                  # {'a': 1, 'c': 3}
```

```python
# clear
d.clear()
print(d)                  # {}
```

```python
# copy
d = {'x': 10}
d2 = d.copy()
print(d2)                 # {'x': 10}
```

```python
# fromkeys
new_dict = dict.fromkeys(['a', 'b', 'c'], 0)
print(new_dict)           # {'a': 0, 'b': 0, 'c': 0}
```

```python
# setdefault
d = {'a': 1}
print(d.setdefault('b', 2))  # 2
print(d)                     # {'a': 1, 'b': 2}
```
