```python
import random

print("성적 처리기 프로그램 ")
print("각 학년마다 반이 1개씩 있고, 반마다 학생 5명이 있습니다.")
print("학생들의 국어, 수학, 영어 점수는 랜덤으로 정해집니다.\n")

# 데이터 만들기
성적표 = {}

for 학년 in range(1, 7):  # 1학년 ~ 6학년
    성적표[str(학년)] = []
    for 학생 in range(5):  # 각 반에 학생 5명
        국어 = random.randint(0, 100)
        수학 = random.randint(0, 100)
        영어 = random.randint(0, 100)
        성적표[str(학년)].append({
            "국어": 국어,
            "수학": 수학,
            "영어": 영어
        })

# 사용자 입력
print("예시: '1학년 국어 평균' 또는 '3학년 수학 1등'")
질문 = input("질문을 입력하세요: ")

단어들 = 질문.split()

if len(단어들) == 3:
    학년 = 단어들[0][0]  # 예: '1학년' -> '1'
    과목 = 단어들[1]      # '국어' 또는 '수학' 또는 '영어'
    종류 = 단어들[2]      # '평균' 또는 '1등'

    if 학년 in 성적표 and 과목 in ["국어", "수학", "영어"]:
        점수목록 = []
        for 학생 in 성적표[학년]:
            점수목록.append(학생[과목])

        if 종류 == "평균":
            평균 = sum(점수목록) / len(점수목록)
            print(f"{학년}학년 {과목} 평균 점수는 {평균:.2f}점입니다.")
        elif 종류 == "1등":
            최고점 = max(점수목록)
            print(f"{학년}학년 {과목} 1등 점수는 {최고점}점입니다.")
        else:
            print("지원하지 않는 질문 종류입니다.")
    else:
        print("없는 학년이거나 과목 이름이 잘못되었습니다.")
else:
    print("질문 형식이 올바르지 않아요! 예: '2학년 국어 평균'")
```
