# **SET 내장함수 설명과 예시**


---


add(x) : 집합에 원소 x를 추가합니다. 이미 존재하면 변화 없음.

---



remove(x)	: 집합에서 원소 x를 제거합니다. x가 없으면 KeyError 발생.

---



discard(x)	: 집합에서 원소 x를 제거합니다. x가 없어도 오류가 발생하지 않음.

---



pop()	: 집합에서 임의의 원소를 제거하고 반환합니다. 집합이 비어 있으면 KeyError 발생.

---



clear()	: 집합의 모든 원소를 제거하여 빈 집합으로 만듭니다.

---



copy()	: 집합을 복사하여 새 집합을 반환합니다.

---



union(set2)	: 두 집합의 합집합을 반환합니다. 중복된 원소는 한 번만 포함.

---



update(set2)	: 현재 집합을 두 집합의 합집합으로 업데이트합니다.

---



intersection(set2)	: 두 집합의 교집합(공통 원소)을 반환합니다.

---



intersection_update(set2)	: 현재 집합을 두 집합의 교집합으로 업데이트합니다.

---



difference(set2)	:  현재 집합에서 다른 집합과의 차집합을 반환합니다.

---



difference_update(set2)	: 현재 집합을 다른 집합과의 차집합으로 업데이트합니다.

---



symmetric_difference(set2)	:  두 집합의 대칭 차집합(서로의 고유 원소)을 반환합니다.

---



symmetric_difference_update(set2)	: 현재 집합을 대칭 차집합으로 업데이트합니다.

---



issubset(set2)	: 현재 집합이 다른 집합의 부분 집합인지 확인하여 True 또는 False 반환.

---



issuperset(set2)	: 현재 집합이 다른 집합을 포함하는 상위 집합인지 확인하여 True 또는 False 반환.

---



isdisjoint(set2)	: 두 집합이 공통 원소가 없는지 확인하여 True 또는 False 반환.

---



len(s)	: 집합의 원소 개수를 반환합니다.

---



in	: 특정 원소가 집합에 포함되어 있는지 여부를 반환합니다.

---



set(iterable)	: 리스트, 튜플 등 반복 가능한 객체를 집합으로 변환합니다. 중복 원소는 제거됩니다.


```python
# add(x) - 원소 추가
s = {1, 2}
s.add(3)
print(s)  # {1, 2, 3}
```

```python
# remove(x) - 원소 제거 (없으면 오류)
s = {1, 2, 3}
s.remove(2)
print(s)  # {1, 3}
```

```python
# discard(x) - 원소 제거 (없어도 오류 없음)
s = {1, 2, 3}
s.discard(4)
print(s)  # {1, 2, 3}
```

```python
# pop() - 임의 원소 제거 후 반환
s = {1, 2, 3}
x = s.pop()
print(x, s)  # 예: 1 {2, 3}
```

```python
# clear() - 모든 원소 제거
s = {1, 2, 3}
s.clear()
print(s)  # set()
```

```python
# copy() - 집합 복사
s = {1, 2, 3}
s2 = s.copy()
print(s2)  # {1, 2, 3}
```

```python
# union(set2) - 합집합 반환
s1 = {1, 2}
s2 = {2, 3}
print(s1.union(s2))  # {1, 2, 3}
```

```python
# update(set2) - 합집합으로 갱신
s1 = {1, 2}
s1.update({2, 3})
print(s1)  # {1, 2, 3}
```

```python
# intersection(set2) - 교집합 반환
s1 = {1, 2, 3}
s2 = {2, 3, 4}
print(s1.intersection(s2))  # {2, 3}
```

```python
# intersection_update(set2) - 교집합으로 갱신
s1 = {1, 2, 3}
s1.intersection_update({2, 3, 4})
print(s1)  # {2, 3}
```

```python
# difference(set2) - 차집합 반환
s1 = {1, 2, 3}
s2 = {2, 3}
print(s1.difference(s2))  # {1}
```

```python
# difference_update(set2) - 차집합으로 갱신
s1 = {1, 2, 3}
s1.difference_update({2, 3})
print(s1)  # {1}
```

```python
# symmetric_difference(set2) - 대칭 차집합 반환
s1 = {1, 2, 3}
s2 = {2, 3, 4}
print(s1.symmetric_difference(s2))  # {1, 4}
```

```python
# symmetric_difference_update(set2) - 대칭 차집합으로 갱신
s1 = {1, 2, 3}
s1.symmetric_difference_update({2, 3, 4})
print(s1)  # {1, 4}
```

```python
# issubset(set2) - 부분 집합 여부 확인
s1 = {1, 2}
s2 = {1, 2, 3}
print(s1.issubset(s2))  # True
```

```python
# issuperset(set2) - 상위 집합 여부 확인
s1 = {1, 2, 3}
s2 = {1, 2}
print(s1.issuperset(s2))  # True
```

```python
# isdisjoint(set2) - 공통 원소 없는지 확인
s1 = {1, 2}
s2 = {3, 4}
print(s1.isdisjoint(s2))  # True
```

```python
# len(s) - 원소 개수 반환
s = {1, 2, 3}
print(len(s))  # 3
```

```python
# in - 원소 포함 여부 확인
s = {1, 2, 3}
print(2 in s)  # True
```

```python
# set(iterable) - 리스트 등을 집합으로 변환
lst = [1, 2, 2, 3]
s = set(lst)
print(s)  # {1, 2, 3}
```
