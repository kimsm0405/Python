```python
def check_length(pw):
    if len(pw) >= 8:
        return 2, " 길이 8자 이상 (+2점)"
    else:
        return 0, " 길이 8자 미만 (+0점)"

def check_digit(pw):
    for ch in pw:
        if ch.isdigit():
            return 2, " 숫자 포함 (+2점)"
    return 0, " 숫자 없음 (+0점)"

def check_upper(pw):
    for ch in pw:
        if ch.isupper():
            return 1, " 대문자 포함 (+1점)"
    return 0, " 대문자 없음 (+0점)"

def check_lower(pw):
    for ch in pw:
        if ch.islower():
            return 1, " 소문자 포함 (+1점)"
    return 0, " 소문자 없음 (+0점)"


check_functions = [check_length, check_digit, check_upper, check_lower]


password = input("비밀번호를 입력하세요: ")


score = 0
feedback_list = []


for func in check_functions:
    s, msg = func(password)
    score += s
    feedback_list.append(msg)


print()
for msg in feedback_list:
    print(msg)


print(f"\n총점: {score}점")


if score >= 6:
    print("비밀번호 강도: 강함")
elif score >= 4:
    print("비밀번호 강도: 보통")
else:
    print("비밀번호 강도: 약함")
```
