```python
def dutch_pay():
    people = ["김", "고", "장", "백"]
    total = {name: 0 for name in people}

    food_cost = int(input("식당 금액을 입력하세요: "))
    food_people = input("식당에 참여한 사람들 (쉼표로 구분): ").split(",")
    share = food_cost // len(food_people)
    for person in food_people:
        person = person.strip()
        if person in total:
            total[person] += share

    bar_cost = int(input("바 금액을 입력하세요: "))
    bar_people = input("바에 참여한 사람들 (쉼표로 구분): ").split(",")
    share = bar_cost // len(bar_people)
    for person in bar_people:
        person = person.strip()
        if person in total:
            total[person] += share

    cafe_cost = int(input("카페 금액을 입력하세요: "))
    cafe_people = input("카페에 참여한 사람들 (쉼표로 구분): ").split(",")
    share = cafe_cost // len(cafe_people)
    for person in cafe_people:
        person = person.strip()
        if person in total:
            total[person] += share

    print("\n개인별 정산 금액:")
    for name in people:
        print(f"{name}: {total[name]}원")

dutch_pay()
```
