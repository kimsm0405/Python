```python
def order_drinks(menu):
    orders = []
    while True:
        item = input("메뉴를 입력하세요 (끝을 입력하면 종료): ")
        if item == "끝":
            break
        if item in menu:
            orders.append(item)
        else:
            print("메뉴에 없습니다. 다시 입력해주세요.")
    return orders

def calculate_total(menu, orders):
    total = 0
    for drink in orders:
        total += menu[drink]
    return total

menu = {
    "아메리카노": 1500,
    "카페라떼": 1800,
    "초코": 2000
}

orders = order_drinks(menu)
total_price = calculate_total(menu, orders)
print("총 결제 금액은", total_price, "원입니다.")
```
