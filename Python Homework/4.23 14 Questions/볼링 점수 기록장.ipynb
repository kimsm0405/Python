```python
def record_bowling_score():
    name = input("이름을 입력하세요: ")
    scores = []

    for i in range(1, 11):
        while True:
            try:
                score = int(input(f"{i}번째 프레임 점수 입력: "))
                if 0 <= score <= 10:
                    scores.append(score)
                    break
                else:
                    print("점수는 0에서 10 사이여야 합니다.")
            except ValueError:
                print("숫자를 입력해주세요.")

    total = sum(scores)

    filename = name + ".txt"
    with open(filename, "w", encoding="utf-8") as f:
        f.write(f"{name}의 볼링 점수\n")
        f.write("--------------------\n")
        for i, s in enumerate(scores, 1):
            f.write(f"{i}프레임: {s}\n")
        f.write(f"\n총점: {total}\n")

    print(f"\n{name}님의 점수가 '{filename}' 파일에 저장되었습니다.")

record_bowling_score()
```
