```python
import random

class Player:
    def __init__(self, name):
        self.name = name
        self.frames = []  # [(first, second, bonus)]
        self.total_score = 0

    def play_game(self):
        for i in range(10):
            first = random.randint(0, 10)
            if first == 10:
                second = 0
            else:
                second = random.randint(0, 10 - first)
            self.frames.append([first, second])

        # 마지막 프레임 보너스 처리
        if self.frames[9][0] == 10 or sum(self.frames[9]) == 10:
            bonus = random.randint(0, 10)
            self.frames[9].append(bonus)
        else:
            self.frames[9].append(0)
```

```python
    def calculate_score(self):
        score = 0
        for i in range(10):
            frame = self.frames[i]
            if frame[0] == 10:  # 스트라이크
                if i < 9:
                    next_frame = self.frames[i + 1]
                    score += 10 + next_frame[0] + (next_frame[1] if next_frame[0] != 10 else self.frames[i + 2][0] if i + 2 < 10 else 0)
                else:
                    score += 10 + frame[2]
            elif sum(frame[:2]) == 10:  # 스페어
                if i < 9:
                    score += 10 + self.frames[i + 1][0]
                else:
                    score += 10 + frame[2]
            else:
                score += sum(frame[:2])
        self.total_score = score

    def show_score(self):
        print(f"\n{self.name}의 점수:")
        for i, frame in enumerate(self.frames):
            print(f"프레임 {i+1}: {frame[:2]}", end='')
            if i == 9 and frame[2] > 0:
                print(f" + 보너스: {frame[2]}", end='')
            print()
        print(f">>> 총점: {self.total_score}\n")
```

```python
class BowlingGame:
    def __init__(self):
        self.players = []

    def start_game(self):
        for i in range(3):
            name = input(f"{i+1}번 플레이어의 이름을 입력하세요: ")
            player = Player(name)
            self.players.append(player)

        for player in self.players:
            player.play_game()
            player.calculate_score()

        for player in self.players:
            player.show_score()

        winner = max(self.players, key=lambda p: p.total_score)
        print(f" 최종 우승자는 {winner.name}입니다! 총점: {winner.total_score}")
```

```python
# 게임 실행
game = BowlingGame()
game.start_game()
```
