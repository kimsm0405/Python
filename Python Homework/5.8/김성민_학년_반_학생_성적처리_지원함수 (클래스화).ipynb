```python
import random

class Student:
    def __init__(self, 과목명):
        self.scores = {과목: random.randint(0, 100) for 과목 in 과목명}

    def 평균(self):
        return sum(self.scores.values()) / len(self.scores)

class ClassRoom:
    def __init__(self, 과목명, 학생수=3):
        self.students = [Student(과목명) for _ in range(학생수)]
        self.과목명 = 과목명

    def 과목별_평균(self):
        과목별_합 = {과목: 0 for 과목 in self.과목명}
        for student in self.students:
            for 과목, 점수 in student.scores.items():
                과목별_합[과목] += 점수

        학생수 = len(self.students)
        return {과목: 합 / 학생수 for 과목, 합 in 과목별_합.items()}

    def 학생별_평균(self):
        return [student.평균() for student in self.students]

class SchoolGradeManager:
    def __init__(self, 학년_수=6, 과목명=None):
        if 과목명 is None:
            과목명 = ["국어", "수학", "영어"]
        self.과목명 = 과목명
        self.학년_수 = 학년_수
        self.학년_목록 = [f"{i+1}학년" for i in range(학년_수)]
        self.학년별_반_수 = [random.randint(3, 10) for _ in range(학년_수)]
        self.school = {}
        self._generate_school()

    def _generate_school(self):
        for 학년_idx, 학년 in enumerate(self.학년_목록):
            self.school[학년] = {}
            반_수 = self.학년별_반_수[학년_idx]
            for 반 in range(반_수):
                반명 = f"{반+1}반"
                self.school[학년][반명] = ClassRoom(self.과목명, 학생수=3)

    def 학년_과목_평균(self, 학년, 과목):
        총합 = 0
        총학생수 = 0
        for 반 in self.school[학년].values():
            for student in 반.students:
                총합 += student.scores[과목]
                총학생수 += 1
        return 총합 / 총학생수 if 총학생수 else 0

    def 학년_반_과목별_평균(self, 학년, 반명):
        반 = self.school[학년][반명]
        return 반.과목별_평균()

    def 학년_과목별_최고점수(self, 학년):
        최고점수 = {과목: 0 for 과목 in self.과목명}
        for 반 in self.school[학년].values():
            for student in 반.students:
                for 과목, 점수 in student.scores.items():
                    if 점수 > 최고점수[과목]:
                        최고점수[과목] = 점수
        return 최고점수

    def 학년_반_학생별_평균점수(self, 학년, 반명):
        반 = self.school[학년][반명]
        return 반.학생별_평균()

    def 학년_석차_평균점수기준(self, 학년):
        평균점수_리스트 = []
        for 반 in self.school[학년].values():
            평균점수_리스트.extend(반.학생별_평균())
        평균점수_리스트.sort(reverse=True)
        return 평균점수_리스트

# 사용 예시
manager = SchoolGradeManager()

print(manager.school)  # 전체 데이터 확인
print("1학년 국어 평균:", manager.학년_과목_평균('1학년', '국어'))
print("1학년 1반 과목별 평균:", manager.학년_반_과목별_평균('1학년', '1반'))
print("1학년 과목별 최고점수:", manager.학년_과목별_최고점수('1학년'))
print("1학년 1반 학생별 평균점수:", manager.학년_반_학생별_평균점수('1학년', '1반'))
print("1학년 평균점수 기준 석차:", manager.학년_석차_평균점수기준('1학년'))
```
