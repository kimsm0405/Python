```python
import random

class RegionPopulation:
    def __init__(self):
        self.regions = [
            "서울", "부산", "대구", "인천", "광주", "대전", "울산",
            "경기", "강원", "충북", "충남", "전북", "전남", "경북", "경남", "제주"
        ]
        self.years = [1970, 1980, 1990, 2000, 2010, 2020]
        self.population_data = {region: {} for region in self.regions}

    def generate_population(self):
        for region in self.regions:
            for year in self.years:
                # 랜덤 인구수 생성 (10만 ~ 1000만 사이)
                self.population_data[region][year] = random.randint(100000, 10000000)

    def print_population(self):
        for region, data in self.population_data.items():
            print(f"지역: {region}")
            for year in self.years:
                print(f"({year}년 : {data[year]:,}명)", end=' ')
            print("\n" + "-"*60)

# 프로그램 실행
region_population = RegionPopulation()
region_population.generate_population()
region_population.print_population()
```
