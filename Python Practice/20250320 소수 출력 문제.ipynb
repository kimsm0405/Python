* 문제
키보드로 숫자를 입력 받음(가정 : 몇 번 입력하는지 모름)
숫자 소수(약수 1과 자기자신) 카운팅
소수가 아니면 다시 입력받음
소수이면 하나씩 카운트 - 소수가 5개가 되면 종료, 소수 화면 출력

```python
p_list = [None, None, None, None, None]   # 값의 개수 5개로 설정, 그 값은 미정
p_index = 0   #index의 0이 가리키는 것은 '첫번째' 값.

while p_index < 5 :   #list에 5개가 채워졌는지 아닌지 확인.
    num = int(input("정수를 입력해주세요: "))
    num_c = 0
    for a in range (1, num + 1):
      if num % a == 0:
          num_c = num_c + 1

    if num_c == 2:
      p_list[p_index] = num
      p_index = p_index + 1

print("소수: ", p_list)
```

```python
def p_check(num):
    num_c = 0
    for a in range(1, num + 1):
        if num % a == 0:
          num_c = num_c + 1

    if num_c == 2:
        return True
    else:
        return False

```

```python
p_list = [None, None, None, None, None]
p_index = 0

while p_index < 5 :
    num = int(input("정수를 입력해주세요: "))

    if p_check(num):
      p_list[p_index] = num
      p_list = p_index + 1


print("소수 : ", p_list)
```
