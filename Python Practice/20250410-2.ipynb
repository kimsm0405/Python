```python
print(" 볼링 점수 계산기 \n")
print("총 10프레임입니다. 각 프레임마다 점수를 직접 입력하세요.\n")

점수들 = []  # 전체 점수를 저장할 리스트




프레임 = 1
while 프레임 <= 10:
    print(f"{프레임}프레임입니다.")
    첫번째 = int(input("첫 번째 투구 점수 (0~10): "))

    if 첫번째 == 10 and 프레임 != 10:
        print("스트라이크입니다!\n")
        점수들.append(10)
        점수들.append(0)  # 두 번째는 생략
        프레임 += 1
    else:
        두번째 = int(input("두 번째 투구 점수 (0~{}): ".format(10 - 첫번째)))
        점수들.append(첫번째)
        점수들.append(두번째)
        프레임 += 1

# 마지막 프레임 보너스 투구
보너스1 = 0
보너스2 = 0

마지막1 = 점수들[-2]
마지막2 = 점수들[-1]

if 마지막1 == 10:
    print("\n마지막 프레임 스트라이크! 보너스 투구 2번 입력하세요.")
    보너스1 = int(input("보너스 첫 투구 (0~10): "))
    보너스2 = int(input("보너스 두 번째 투구 (0~10): "))
elif 마지막1 + 마지막2 == 10:
    print("\n마지막 프레임 스페어! 보너스 투구 1번 입력하세요.")
    보너스1 = int(input("보너스 투구 (0~10): "))

점수들.append(보너스1)
점수들.append(보너스2)

# 총합 계산
총점 = 0
i = 0
프레임 = 1

while 프레임 <= 10:
    첫 = 점수들[i]
    두 = 점수들[i + 1]

    if 첫 == 10:
        보너스a = 점수들[i + 2]
        보너스b = 점수들[i + 3]
        총점 += 10 + 보너스a + 보너스b
        i += 1
    elif 첫 + 두 == 10:
        보너스a = 점수들[i + 2]
        총점 += 10 + 보너스a
        i += 2
    else:
        총점 += 첫 + 두
        i += 2
    프레임 += 1

print("\n 게임 종료! 당신의 총점은:", 총점, "점입니다!")
```
